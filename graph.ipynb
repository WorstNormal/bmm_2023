```python



with open("./output/mse.txt", "r") as f:
    mse_array = [float(x) for x in f.read().split("\n")[:-1][:100]]
    ax = sns.lineplot(x=[i for i in range(len(mse_array))], y=mse_array)
    ax.set(xlabel="Simulation number", ylabel="MSE")
```
